```python
df
```

```python
new_entry={'name':'raju','masters':'MK','age':26}
df=df.append(new_entry,ignore_index=True)
df
```

```python
new_entry={'name':'raju1','masters':'MK1','age':261}
df=df.append(new_entry,ignore_index=False)
df
```

```python
df.sort_index(ascending=True,inplace=True)
df
```
